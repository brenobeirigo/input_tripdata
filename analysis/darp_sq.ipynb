```python
import pandas as pd
import os
import sys

# Import results file to the same folder
df = pd.read_csv("results_paper.csv")
df.columns
```

```python
#filter_cols = ['test', 'demand_size',
#       'user_base', 'id_instance', 'fleet_capacity_obj_val', 'total_pk_delay', 'total_ride_delay',
#       'total_delay_obj_val',   'tier_1', 'tier_2', 'tier_1_pk_delay_sum', 'tier_2_pk_delay_sum', 'pk_delay_mean_A', 'pk_delay_mean_B', 'pk_delay_mean_C', 'run_time']

df["fleet_size"] = df["fleet_capacity_obj_val"]/df["max_vehicle_capacity"]
# filter_cols = ['test', 'demand_size', 'user_base', 'id_instance', 'fleet_size', "total_pk_delay_A", "total_pk_delay_B", "total_pk_delay_C",  'total_pk_delay', 'total_ride_delay','run_time']
filter_cols = ['test', 'demand_size', 'user_base', 'id_instance', 'fleet_size',  'total_pk_delay', 'total_ride_delay','run_time']

dff = df[filter_cols]
# filter_runtime = dff["run_time"]<1600
# dff=dff[filter_runtime]
dff.sort_values(by=['demand_size', 'user_base', 'test', 'id_instance', 'run_time'])
```

```python
dfff =dff.loc[dff["demand_size"]<15]
dfff = dfff.rename(columns={"demand_size":"#R", "run_time":"Runtime(s)", "user_base":"User base", "test": "Service quality", "fleet_size": "#K", "total_pk_delay": "Pickup","total_pk_delay_A": "Business","total_pk_delay_B": "Standard","total_pk_delay_C": "Low-cost", "total_ride_delay": "Ride delay(s)"})
dfff
```

```python
dfff = dfff.groupby(["#R", "User base", "Service quality"]).mean()
dfff = dfff.sort_values(by=["#R", "User base", "Service quality"])
dfff
```

```python
dfff = dfff.reset_index()
sq_labels = dict(baseline_1="SL1", slevels="SQA", baseline_2="SL2")
u_base = dict(AA="B+", BB="S+", CC="L+")
sqs = pd.api.types.CategoricalDtype(categories=["SL1", "SQA", "SL2"], ordered=True)

dfff["Service quality"] = dfff["Service quality"].apply(lambda x:sq_labels[x])
dfff["User base"] = dfff["User base"].apply(lambda x:u_base[x])

dfff["Service quality"] = dfff["Service quality"].astype(sqs)
dfff= dfff.drop(["id_instance"], axis=1)
dfff = dfff.sort_values(by=["#R", "User base", "Service quality"])
def two(x):
    print(x, f"{x:.2f}")
    if x:
        return float(f"{x:.2f}")
    else:
        return "-"
dfff["Runtime(s)"]=dfff["Runtime(s)"].astype(str)
dfff["Runtime(s)"] = dfff["Runtime(s)"].apply(lambda x:two(float(x)))
dfff = dfff.groupby(["#R", "User base", "Service quality"]).mean()
dfff=dfff.round(1)
```

```python
dffff = dfff.astype(str)
dffff.to_csv("aggregated_results.csv")
dffff
```
