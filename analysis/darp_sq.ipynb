```python
import pandas as pd

# Import results file to the same folder
df = pd.read_csv("results.csv")
# NOTE: The random class distribution sometimes don't assign all classes (you find  blanks '' that become Nan)
print(f"#Entries: {len(df.columns)}")
df.columns
max_requests = 20
max_runtime = 18000
index_table = [
    "#R",
    "User base",
    "Service quality"
]
```

```python
columns = {
    "test":"Service quality",
    "id_instance": "Id",
    "demand_size":"#R",
    "user_base":"User base",
    "fleet_size": "#K",
    "pk_delay_mean_A": "Business(pk)",
    "pk_delay_mean_B": "Standard(pk)",
    "pk_delay_mean_C": "Low-cost(pk)",
    #"total_pk_delay_A": "Business(total)",
    #"total_pk_delay_B": "Standard(total)",
    #"total_pk_delay_C": "Low-cost(total)",
    "ride_delay_mean_A": "Business(ride)",
    "ride_delay_mean_B": "Standard(ride)",
    "ride_delay_mean_C": "Low-cost(ride)",
    #"total_A": "Business(total)",
    #"total_B": "Standard(total)",
    #"total_C": "Low-cost(total)",
    "total_pk_delay": "Pickup",
    "total_ride_delay": "Ride delay(s)",
    #"total_delay": "Total delay(s)",
    "run_time":"Runtime(s)"
}

sq_labels = dict(
    status_quo="0%",
    #slevels_050="50%",
    slevels_080="80%",
    slevels_090="90%",
    slevels_100="100%"
)

u_base = dict(
    AA="B+",
    BB="S+",
    CC="L+",
    #B="B"
)


df = df[df["user_base"].isin(set(u_base.keys()))]
df = df[df["test"].isin(set(sq_labels.keys()))]
```

```python
df["fleet_size"] = df["fleet_capacity_obj_val"]/df["max_vehicle_capacity"]

filter_cols = list(columns.keys())

dff = df[filter_cols]

dff
```

```python
dff = dff.sort_values(by=['demand_size', 'user_base', 'test', 'id_instance', 'run_time'])

dff
```

## Filter max. number of requests

```python
print(f"Size before filtering: {len(dff)}")
dfff =dff.loc[dff["demand_size"]<=max_requests]
print(f"Size AFTER filtering: {len(dfff)}")

dfff.head(50)
```

```python
df_maxruntime = dfff.copy()
print(f"Size before filtering: {len(dfff)}")
df_maxruntime =df_maxruntime.loc[df_maxruntime["run_time"]>=max_runtime]
print(f"Size AFTER filtering: {len(df_maxruntime)}")
```

## Filter max. run time similar instances
For the sake of comparison, we eliminate instances that present the same characteristics (id_instance, demand_size, user_base) of the run_time excluded instances.
This way, the comparison between average instances is fairer.

```python
df_clean = dff.copy()
```

```python
df_clean
```

```python
for i, r in df_maxruntime.iterrows():
    filter = ~((df_clean["id_instance"]==r["id_instance"])&(df_clean["demand_size"]==r["demand_size"])&(df_clean["user_base"]==r["user_base"]))
    df_clean = df_clean.loc[filter]

dfff = df_clean
dfff
```

## Renaming columns

```python
dfff = dfff.rename(columns=columns)
dfff.head(50)
```

```python
counts = dfff[["#R", "User base", "Service quality", "Id"]].groupby(["#R", "User base", "Service quality"]).agg(["count"])
counts
```

```python
dfff = dfff.groupby(["#R", "User base", "Service quality"]).mean()
print(dfff.columns)
dfff = dfff.sort_values(by=["#R", "User base", "Service quality"])


dfff
```

```python
dfff = dfff.reset_index()

sqs = pd.api.types.CategoricalDtype(categories=list(sq_labels.values()), ordered=True)
ubases = pd.api.types.CategoricalDtype(categories=list(u_base.values()), ordered=True)

dfff["Service quality"] = dfff["Service quality"].apply(lambda x:sq_labels[x])
dfff["User base"] = dfff["User base"].apply(lambda x:u_base[x])

dfff["User base"] = dfff["User base"].astype(ubases)
dfff["Service quality"] = dfff["Service quality"].astype(sqs)
```

```python
columns_to_remove = [
    #"#R",
    "Id"
]

dfff = dfff.drop(columns_to_remove, axis=1)
dfff
```

```python
# Blank cols to create latex
dfff["space_col_1"] = ""
dfff["space_col_2"] = ""
dfff["space_col_3"] = ""

df_added_blank_cols = dfff[
    [
        "#R",
        "User base",
        "Service quality",
        "#K", "space_col_1",
        "Business(pk)",
        "Standard(pk)",
        "Low-cost(pk)",
        "space_col_2",
        "Business(ride)",
        "Standard(ride)",
        "Low-cost(ride)",
        "space_col_3",
        #"Pickup",
        #"Ride delay(s)",
        "Runtime(s)"
    ]
]
df_added_blank_cols
```

```python
def two(x):
    print(x, f"{x:.2f}")
    if x:
        return float(f"{x:.2f}")
    else:
        return "-"

# Only runtime is has two cases
df_added_blank_cols["Runtime(s)"]=df_added_blank_cols["Runtime(s)"].astype(str)
df_added_blank_cols = df_added_blank_cols.round(1)
df_added_blank_cols["Runtime(s)"] = df_added_blank_cols["Runtime(s)"].apply(lambda x:two(float(x)))

df_grouped = df_added_blank_cols.sort_values(by=index_table)

df_grouped
```

```python
df = df_grouped.set_index(index_table)
df
```

```python
print(df.columns)
df["Runtime(s)"] = counts.values
```

```python
df = df.astype(str)
df.to_csv("aggregated_results.csv")
df
```
